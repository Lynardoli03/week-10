```python
#Problem 10 programming exercises




class Pet:
    def __init__(self, name="", animal_type="", age=0):    #setting the inputs equal to eachother so they return the same thing
        self.__name = name
        self.__animal_type = animal_type
        self.__age = age

    # This is essentially the questions I ant asked and what it will output back
    def set_name(self, name):
        self.__name = name

    def set_animal_type(self, animal_type):
        self.__animal_type = animal_type

    def set_age(self, age):
        self.__age = age

    # It will input what they typed and spit it back out
    def get_name(self):
        return self.__name

    def get_animal_type(self):
        return self.__animal_type

    def get_age(self):
        return self.__age

# This is the Main code that will run the program
def main():
    # I am making/setting up the scenario
    my_pet = Pet()
    
    # Get user input for the pet's details
    name = input("Enter your pet's name: ")
    animal_type = input("Enter the type of your pet (e.g., Dog, Cat, Bird): ")
    age = int(input("Enter your pet's age: "))
    
    # setting the attributes of the pet
    my_pet.set_name(name)
    my_pet.set_animal_type(animal_type)
    my_pet.set_age(age)
    
    # Retrieve and display the pet's attributes 
    print("\nPet's Details:")
    print("Name:", my_pet.get_name())
    print("Type:", my_pet.get_animal_type())
    print("Age:", my_pet.get_age())

if __name__ == "__main__":
    main()
```

```python

#Problem 3 programming exercises. 


class PersonalInformation:
    def __init__(self, name="", address="", age=0, phone_number=""):
        self.__name = name
        self.__address = address
        self.__age = age
        self.__phone_number = phone_number

    # the questions the code is asking
    def set_name(self, name):
        self.__name = name

    def set_address(self, address):
        self.__address = address

    def set_age(self, age):
        self.__age = age

    def set_phone_number(self, phone_number):
        self.__phone_number = phone_number

    # the answers that it will return
    def get_name(self):
        return self.__name

    def get_address(self):
        return self.__address

    def get_age(self):
        return self.__age

    def get_phone_number(self):
        return self.__phone_number

# Main program that uses the code
def main():
    # Created three instances of PersonalInformation with 
    person1 = PersonalInformation("John Doe", "123 Elm Street", 30, "555-0101")
    person2 = PersonalInformation("Jane Smith", "456 Oak Avenue", 28, "555-0202")
    person3 = PersonalInformation("Alice Johnson", "789 Pine Road", 35, "555-0303")

    # list of people to go through
    persons = [person1, person2, person3]

    # Display each person's information
    for person in persons:
        print("Name:", person.get_name())
        print("Address:", person.get_address())
        print("Age:", person.get_age())
        print("Phone Number:", person.get_phone_number())
        print()  # Print a newline 

if __name__ == "__main__":
    main()
```
